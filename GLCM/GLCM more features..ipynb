```python
import os
import cv2
import numpy as np
import mahotas.features.texture as mhtex
from skimage.feature import graycomatrix, graycoprops


folder_path = 'carcinoma_in_situ'

for filename in os.listdir(folder_path):
    if filename.endswith('.png'):
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path, 0)  # Read the image in grayscale

        distances = [1]
        angles = [0, np.pi / 4, np.pi / 2, 3 * np.pi / 4]

        glcm = graycomatrix(image, distances, angles, levels=256, symmetric=True, normed=True)

        properties = ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation']
        features = np.ravel([graycoprops(glcm, prop).reshape(-1) for prop in properties])

        # Additional properties using mahotas
        haralick_features = mhtex.haralick(image)
        asm = haralick_features[:, 0].mean()
        cluster_prominence = haralick_features[:, 3].mean()
        cluster_shade = haralick_features[:, 4].mean()
        max_prob = haralick_features[:, 8].mean()

        print(f'Features for image: {filename}')
        for prop, feature in zip(properties + ['ASM', 'cluster_prominence', 'cluster_shade', 'max_prob'], features.tolist() + [asm, cluster_prominence, cluster_shade, max_prob]):
            print(f'{prop}: {feature}')
        print('------------------------------')
```

```python

                    # GLCM More Fetures added
# 'contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation',
# 'ASM', 'cluster_prominence', 'cluster_shade', 'max_prob','LBP','HOG'.

import os
import cv2
import numpy as np
from skimage.feature import greycomatrix, graycoprops, local_binary_pattern, hog
import mahotas.features.texture as mhtex

folder_path = 'carcinoma_in_situ'

for filename in os.listdir(folder_path):
    if filename.endswith('.png'):
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path, 0)  # Read the image in grayscale

        distances = [1]
        angles = [0, np.pi / 4, np.pi / 2, 3 * np.pi / 4]

        glcm = graycomatrix(image, distances, angles, levels=256, symmetric=True, normed=True)

        properties = ['contrast', 'dissimilarity', 'homogeneity', 'energy', 'correlation']
        features = np.ravel([graycoprops(glcm, prop).reshape(-1) for prop in properties])

        # Additional properties using mahotas
        haralick_features = mhtex.haralick(image)
        asm = haralick_features[:, 0].mean()
        cluster_prominence = haralick_features[:, 3].mean()
        cluster_shade = haralick_features[:, 4].mean()
        max_prob = haralick_features[:, 8].mean()

        # Additional features using skimage
        radius = 3
        n_points = 8 * radius
        lbp = local_binary_pattern(image, n_points, radius, method='uniform')
        lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), range=(0, n_points + 2))
        lbp_hist = lbp_hist.astype(np.float)
        lbp_hist /= (lbp_hist.sum() + 1e-6)  # Normalize LBP histogram

        hog_features = hog(image, orientations=8, pixels_per_cell=(16, 16), cells_per_block=(1, 1))

        print(f'Features for image: {filename}')
        for prop, feature in zip(properties + ['ASM', 'cluster_prominence', 'cluster_shade', 'max_prob'],
                                 features.tolist() + [asm, cluster_prominence, cluster_shade, max_prob]):
            print(f'{prop}: {feature}')
        print(f'LBP: {lbp_hist.tolist()}')
        print(f'HOG: {hog_features.tolist()}')
        print('------------------------------')
```

```python
import os
import glob
import numpy as np
from skimage.feature import greycomatrix, greycoprops
from skimage import io, color, img_as_ubyte

# Replace 'folder_path' with the path to the folder containing your PNG images
folder_path = 'carcinoma_in_situ'

# List all PNG image files in the folder
image_files = glob.glob(os.path.join(folder_path, '*.png'))

# Function to compute texture features for a single image
def compute_texture_features(image_path):
    image = io.imread(image_path)
    # Convert the image to unsigned 8-bit integer (uint8)
    gray_image = img_as_ubyte(color.rgb2gray(image))
    
    distances = [1, 2, 3]  # You can adjust the distances as needed
    angles = [0, np.pi/4, np.pi/2, 3*np.pi/4]  # You can adjust the angles as needed
    levels = 256  # Number of gray levels, usually 256 for 8-bit images
    glcm = greycomatrix(gray_image, distances=distances, angles=angles, levels=levels, symmetric=True, normed=True)
    
    features = {
        'ASM': greycoprops(glcm, 'ASM').ravel(),
        'Contrast': greycoprops(glcm, 'contrast').ravel(),
        'Correlation': greycoprops(glcm, 'correlation').ravel(),
        'Homogeneity': greycoprops(glcm, 'homogeneity').ravel(),
        'Energy': greycoprops(glcm, 'energy').ravel(),
        'Dissimilarity': greycoprops(glcm, 'dissimilarity').ravel(),
    }
    
    return features

# Process all images in the folder and print the computed features
for image_file in image_files:
    image_features = compute_texture_features(image_file)
    print(f"Image: {os.path.basename(image_file)}")
    for feature_name, feature_values in image_features.items():
        print(f"{feature_name}: {feature_values}")
    print("\n")
```

```python

                      # Gray scale on carcinoma_situ

import os
import cv2

def convert_to_grayscale(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename)

        # Read the image
        image = cv2.imread(input_path)

        if image is not None:
            # Check if the image is already grayscale
            if len(image.shape) == 2:  # Grayscale images have only two dimensions (height and width)
                print(f"{filename} is already grayscale. Copying as is.")
                cv2.imwrite(output_path, image)
            else:
                # Convert to grayscale
                gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

                # Save the grayscale image
                cv2.imwrite(output_path, gray_image)

                print(f"{filename} converted to grayscale and saved.")

if __name__ == "__main__":
    input_folder = "carcinoma_in_situ"
    output_folder = "carcinoma_in_situ_change"

    convert_to_grayscale(input_folder, output_folder)
```
